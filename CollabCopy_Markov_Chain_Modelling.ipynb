While we will do our best to minimize the use of technical jargon for our readers, it will not always be possible to do so.  First, we'd like to define some the critical components of isotope decay.
    <br><br>
    What is isotope decay?  Isotopes are atoms of an element with different numbers of nuetrons than are normal present in atoms of that element.  In our project, and example of this would be Strontium-90.  The 90 in the name indicates that we are talking about the element Strontium whose atoms have 38 protons and 52 neutrons, which is 2 more nuetrons than typical Strontium.  Throughout this project we will often refer to isotopes using their chemical symbol and their atomic number; For example Strontium-90 will be Sr-90.  
    <br><br>
    Now that we've discussed what an isotope is, what do we mean by decay?  Decay is referring to the process by which radioactive (unstable) isotopes become more stable isotopes, often isotopes of other elements.  Generally we refer to these elements as the parent isotope and the daughter isotope(s), respectively. A parent typically decays into a single daughter, but there are some isotopes that decay into two daughters through different processes. There are many ways that isotopes can shed their energy via radiation, but we are not particularily concerned about distinguishing them in this project.  We will frequently use the term isotop decay chains, which is how we describe all of the isotopes related in a chain of decays from an isotope parent all the way down to its stable element. One of hte chains we will be considering throughout this project is decay chain Sr-90 to Zr-90, which includes both the parent and granddaughter, but Y-90 as well. Now that we've discussed what isotope decay is, we can explore some of the technical ideas behind modelling it.
    <br><br>
    Isotope decay can be modeled using some very simple parameters which are intricately related: decay rate and half life.  The decay rate is pretty much exactly what it sounds like: It is the rate at which an element decays.  The half life is the amount of time it takes for an isotope to decay half its orginal amount.  Given decay rate you can find half life and vice versa.  We will provide a formula in the Mathematical Formulation section, where their connection can be seen clearly.
    <br><br>
    This analysis will rely heavily on undergraduate level mathmatics, and while we don't have the time or space to teach our readers some of the strategies used, we can take some time now to explain their function in this project.  We will be using four techniques to analyze isotope decay: Solving systems of differential equations with eigenspaces and Laplace transformations, using the Bateman equation, and Markov Chains. While each technique ultimately serves the same purpose of modeling isotope decay, they each have their own advantages and caveats.
    <br><br>
    Both solving the system of differential equations with eigenspaces and Markov processes utilize linear algebra methods to model system behavior.  While very useful, they have considerable limitations as they both become quite unwieldy in chains greater than 5.  The Markov chain relies on probability to model system state changes over time, and evades the issues presented by solving the differential equations.  Laplace transformations are an ordinary differential equations technique that allows for easier solutions for chains greater than 5, and is ultimately how the Bateman equation is derived.  We will finally use the Bateman equations which provide the easiest method of calculating decay rates of any chain greater than 2.  We will explore these ideas in more depth as we utilize them throughout this project. Now that we have defined many of the technical terms, we are ready to proceed to the mathematics.

<h3>
  3-1: Simply Decay Example: Sr-90 to Zc-90
</h3>
<p>
  First we'll start by using eigenspaces to find model the decay chain of  Strontium 90 to Yttrium 90 to Zirconium 90.  Figure 3.1.1 from Periodic Table (Gray) gives a visual representation of the chain.  According to a lecture by Dominic Sims with the International Atomic Energy Agency, Strontium and Yttrium have halflives of 28 years and 64 hours, respectively.  From the same source, our differential equation can be verified to be correct.
</p>
<h5>
  <b>
    Figure 3.1.1 Strontium 90 Decay Chain illustration from Periodic Table (Gray)
  </b>
</h5>


\begin{align*}
    Sr^{90}_0\;{\rightarrow}\;Y^{90}_1\;{\rightarrow}\;Zr^{90}_2\\
\end{align*}

<p>
  Differential Equations:
</p>
\begin{align*}
    \frac{dSr^{90}_0}{dt}=-{\lambda}_0Sr^{90}_0,\quad\frac{dY^{90}_1}{dt}=-{\lambda}_1Y^{90}_1+{\lambda}_0Sr^{90}_0,\quad\frac{dZr^{90}_2}{dt}={\lambda}_1Y^{90}_1 \\
\end{align*}

<p>
  Matrix Equations:
</p>
\begin{align*}
    % Derivative Vector
    \begin{bmatrix}
        Sr'^{90}_{0} \\
        Y'^{90}_{1} \\
        Zr'^{90}_{2} \\
    \end{bmatrix}=
    % Values
    \begin{bmatrix}
        -{\lambda}_0 & 0 & 0 \\
        -{\lambda}_1 & {\lambda}_0 & 0 \\
        {\lambda}_1 & 0 & 0 \\
    \end{bmatrix}
    % Solution Vector
   \begin{bmatrix}
        Sr^{90}_0 \\
        Y^{90}_1 \\
        Zr^{90}_2 \\
    \end{bmatrix}=\;A\textbf{y}=\textbf{y'}\\
\end{align*}

<p>
  Eigenvector Expression to get characteristic equation:
</p>
\begin{align*}
    |A-{\mu}I_3|=\textbf{0}\\
    det
    \begin{bmatrix}
        -{\lambda}_0-{\mu} & 0 & 0 \\
        -{\lambda}_1 & {\lambda}_0-{\mu}  & 0 \\
        {\lambda}_1 & 0 & -{\mu} \\
    \end{bmatrix} = \textbf{0} \\
\end{align*}

<p>
  Characteristic expression:
</p>
\begin{align*}
    \mu({\lambda}_0+{\mu})({\lambda}_0-{\mu})=0 \\
\end{align*}

<p>
  Eigenvalues:
</p>
\begin{align*}
    {\mu}_0=0,\quad{\mu}_{-{\lambda}_0}=-{\lambda}_0,\quad{\mu}_{{\lambda}_0}={\lambda}_0\\
\end{align*}

<p>
  Eigenvalue = 0 substitution:
</p>
\begin{align*}
    {\mu}_0=0:\quad
    \begin{bmatrix}
        -{\lambda}_0 & 0 & 0 \\
        -{\lambda}_1 & {\lambda}_0 & 0 \\
        {\lambda}_1 & 0 & 0 \\
    \end{bmatrix}, \quad
    \textbf{y}_0 = c_0
    \begin{bmatrix}
        0 \\
        0 \\
        1 \\
    \end{bmatrix}\\
\end{align*}

<p>
  Eigenvalue = -lambda substitution:
</p>
\begin{align*}
    {\mu}_{-{\lambda}_0}=-{\lambda}_0:\quad
    \begin{bmatrix}
        -{\lambda}_0-(-{\lambda}_0) & 0 & 0 \\
        -{\lambda}_1 & {\lambda}_0-(-{\lambda}_0) & 0 \\
        {\lambda}_1 & 0 & -(-{\lambda}_0) \\
    \end{bmatrix}{\rightarrow}
    \begin{bmatrix}
        0 & 0 & 0 \\
        -{\lambda}_1 & 2{\lambda}_0 & 0 \\
        {\lambda}_1 & 0 & {\lambda}_0 \\
    \end{bmatrix} \\
    \textbf{y}_{-{\lambda}_0} = c_{-{\lambda}_0}
    \begin{bmatrix}
        \frac{{\lambda}_0}{{\lambda}_1} \\
        \frac{1}{2} \\
        -1 \\
    \end{bmatrix}e^{-{\lambda}_0t}\\
\end{align*}

<p>
  Eigenvalue = lambda substitution:
</p>
\begin{align*}
    {\mu}_{{\lambda}_0}={\lambda}_0:\quad
    \begin{bmatrix}
        -{\lambda}_0-{\lambda}_0 & 0 & 0 \\
        -{\lambda}_1 & {\lambda}_0-{\lambda}_0 & 0 \\
        {\lambda}_1 & 0 & -{\lambda}_0 \\
    \end{bmatrix}{\rightarrow}
    \begin{bmatrix}
        -2{\lambda}_0 & 0 & 0 \\
        -{\lambda}_1 & 0 & 0 \\
        {\lambda}_1 & 0 & -{\lambda}_0 \\
    \end{bmatrix} \\
    \textbf{y}_{{\lambda}_0} = c_{{\lambda}_0}
    \begin{bmatrix}
        0 \\
        1 \\
        0 \\
    \end{bmatrix}e^{{\lambda}_0t}\\
\end{align*}

<p>
  General Solution:
</p>
\begin{align*}
    \textbf{y}=c_0
    \begin{bmatrix}
        0 \\
        0 \\
        1 \\
    \end{bmatrix}+c_{-{\lambda}_0}
    \begin{bmatrix}
        \frac{{\lambda}_0}{{\lambda}_1} \\
        \frac{1}{2} \\
        -1 \\
    \end{bmatrix}e^{-{\lambda}_0t}+c_{{\lambda}_0}
    \begin{bmatrix}
        0 \\
        1 \\
        0 \\
    \end{bmatrix}e^{{\lambda}_0t}\\
\end{align*}

<p>
  IVP:
</p>
\begin{align*}
    \textbf{y}(0)=
    \begin{bmatrix}
        Sr^{90}_0 \\
        0 \\
        0 \\
    \end{bmatrix}{\rightarrow}
    \begin{bmatrix}
        0 & \frac{{\lambda}_0}{{\lambda}_1} & 0 & Sr^{90}_0 \\
        1 & \frac{1}{2} & 0 & 0 \\
        0 & -1 & 1 & 0 \\
    \end{bmatrix},\quad c_0=c_{-{\lambda}_0}=Sr^{90}_0\frac{{\lambda}_1}{{\lambda}_0},c_{{\lambda}_0}=-Sr^{90}_0\frac{{\lambda}_1}{2{\lambda}_0} \\
\end{align*}

<p>
  General Solution w/ intial conditions:
</p>
\begin{align*}
    \textbf{y}(t)=Sr^{90}_0\frac{{\lambda}_1}{{\lambda}_0}
    \begin{bmatrix}
        0 \\
        0 \\
        1 \\
    \end{bmatrix}+Sr^{90}_0\frac{{\lambda}_1}{{\lambda}_0}
    \begin{bmatrix}
        \frac{{\lambda}_0}{{\lambda}_1} \\
        \frac{1}{2} \\
        -1 \\
    \end{bmatrix}e^{-{\lambda}_0t}-Sr^{90}_0\frac{{\lambda}_1}{2{\lambda}_0}
    \begin{bmatrix}
        0 \\
        1 \\
        0 \\
    \end{bmatrix}e^{{\lambda}_0t}\\
\end{align*}

<p>
  Final equations:
</p>
\begin{align*}
    \textbf{Sr}^{90}_0(t)=Sr_{90}e^{-{\lambda}_0t}_0,\quad \textbf{Y}^{90}_1(t)=Sr^{90}_0\frac{{\lambda}_1}{2{\lambda}_0}e^{{\lambda}_0t}(e^{-2{\lambda}_0t}-1),\quad\textbf{Zr}^{90}_2(t)=Sr^{90}_0\frac{{\lambda}_1}{{\lambda}_0}(1-e^{-{\lambda}_0t})\\
\end{align*}

<p>
  Half life substitutions:
</p>
\begin{align*}
    {\lambda}_0=\frac{ln(2)}{{\tau}_0},\quad{\lambda}_1=\frac{ln(2)}{{\tau}_1} \\
\end{align*}

<p>

\begin{align*}
    \textbf{Sr}^{90}_0(t)=Sr^{90}_0e^{-\frac{ln(2)}{{\tau}_0}t},\quad\textbf{Y}^{90}_{1}(t)=Sr^{90}_0\frac{{\tau}_0}{2{\tau}_1}(e^{\frac{-2ln(2)}{{\tau}_0}t}-1),\quad\textbf{Zr}^{90}_{2}(t)=Sr^{90}_0\frac{{\tau}_0}{{\tau}_1}(1-e^{-\frac{ln(2)}{{\tau}_0}t})
\end{align*}

```python
import numpy as np
import matplotlib.pyplot as plt

# define a function for the amounts of two isotopes from the model above.
def decay_two_iso(half_life, init_amount, t):

    '''
        Returns the amounts of each isotope with respect to a time.

            Parameters:
                half_life (float): half-life of the initial isotope.
                init_amount (float): How much of the initial isotope we have.
                t (float/array): A float one of the times that we are going to check.

            Returns:
                N1 (float/array): returns a float of the amount for a given time isotope 1.
                N2 (float/array): returns a float of the amount for a given time isotope 2.

    '''

    # Recast the values for ease of typing
    A = init_amount
    hl = half_life

    # Compute the values for the amounts from the above math.
    N1 = A * np.exp((-np.log(2) * t) / hl)
    N2 = -A * np.exp((-np.log(2) * t) / hl) + A

    # Return the computed amounts.
    return N1, N2, t


# Coded in collaboration with Chase Varner and Gavin Miller
# define a function for the amounts of two isotopes from the model above.
def decay_three_iso(half_life_1, half_life_2, init_amount_1, init_amount_2, init_amount_3, t):

    '''
        Returns the amounts of each isotope with respect to a time.

            Parameters:
                half_life_1 (float): half-life of the initial isotope.
                half_life_2 (float): half-life of the second isotope.
                init_amount_1 (float): How much of the first isotope we have.
                init_amount_1 (float): How much of the second isotope we have.
                init_amount_1 (float): How much of the third isotope we have.
                t (float/array): A float one of the times that we are going to check.

            Returns:
                N1 (float/array): returns a float of the amount for a given time isotope 1.
                N2 (float/array): returns a float of the amount for a given time isotope 2.
                N3 (float/array): returns a float of the amount for a given time isotope 3.
                t (float/array): returns the times.

    '''

    # This will convert the half life to a decay rate.
    u1 = np.log(2) / half_life_1
    u2 = np.log(2) / half_life_2

    # This will assign the values to C1, C2 and C3 from the (by hand) solution.
    C3 = init_amount_1
    C2 = init_amount_2 + init_amount_1 * (u1 / (u1-u2))
    C1 = init_amount_3 + init_amount_2 + init_amount_1 * (u1 / (u1 - u2)) - init_amount_1 * ((u2) / (u1 - u2))


    # Compute the values for the amounts from by hand work for the solution to the system
    n1 = C3 * np.exp(-u1 * t)
    n2 = C2 * np.exp(-u2 * t) + C3 * (-u1/(u1 - u2)) * (np.exp(-u1 * t))
    n3 = C1 - C2 * np.exp(-u2 * t) + C3 * ((u2) / (u1 - u2)) * (np.exp(-u1 * t))

    # Return the computed amounts.
    return n1, n2, n3, t

# Create a range of times to send to the above definition.
# The first number is the start time, the second number is
# the end time and the final value is the step size.
times = np.arange(0,100,1)

# Send all the times from the above array into the definition that we have created above.
amounts_over_time = decay_three_iso(25, 3, 1, 0, 0, times)

# This will
print(np.shape(amounts_over_time))

decay_two_iso(10,23,10)

# Next lines until 'plt.show()' was sourced from Forland, Adam, Isotope_Project_Part1.ipynb
# Set the timeline
time = np.arange(0, 5.8 * 10**9, 10**8)

# Plot the functions by passing it the time and the function (with the arguements from above)
plt.plot(time, decay_two_iso(1.28 * 10**9, 10, time)[0], marker = "." , label="K")
plt.plot(time, decay_two_iso(1.28 * 10**9, 10, time)[1], marker = "." ,label="Ca")

# Plot the guidance information.
plt.legend()
plt.title("K --> Ca")
plt.xlabel("Time years")
plt.ylabel("Amount")

# Show the plot.
plt.show()

# Plot each of the three isotopes.
plt.plot(times, amounts_over_time[0], label="iso1")
plt.plot(times, amounts_over_time[1], label="iso2")
plt.plot(times, amounts_over_time[2], label="iso3")

plt.legend()

# Show the plot.
plt.show()
```
